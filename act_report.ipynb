# "WeRateDogs" Analysis

"WeRateDogs" Tweets helped us in finding the most favorate dogs stage, the trendy stage and the highly rated ones. The following charts displays the relation between each of these points.

```python
# Import generally needed libraries across the code
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("darkgrid")
```

```python
# Load the cleaned tweets datatset
df = pd.read_csv('./twitter_archive_master.csv')
```

```python
df['stage'].hist(figsize=(10,5));
plt.xlabel("Dog Stage")
plt.ylabel("No. of Dogs")
plt.title("The Distribution of The Rated Dogs Stages")
plt.xticks(rotation=45);
```

This chart showed the **pupper** stage is the most common stage in the selected sample of the retweets.

```python
stages_ratings = df.groupby('stage')['rating'].mean()
stages_ratings.plot(kind = 'bar', figsize=(10,5), color=(0.2, 0.4, 0.6, 1))
plt.xlabel("Dog Stage")
plt.ylabel("Rating")
plt.title("The Distribution of The Ratings For The Different Dogs Stages");
plt.xticks(rotation=45);
```

Generally, the different dog stages got similar rating. This indicates that the dog stage does not affect much the rating based on the studied dataset. <br>
All the information related to dog stages are for 15.8% of the analyzed dataset. As the dogs stages were given for this portion only.<br>
Specifcally `doggo, puppo` stage got the highest average rating of 130%.

```python
stages_favorite = df.groupby('stage')['favorite_count'].mean()
stages_favorite.plot(kind = 'bar', figsize=(10,5), color=(0.2, 0.4, 0.6, 1))
plt.xlabel("Dog Stage")
plt.ylabel("Favorate Count")
plt.title("The Relation Between Favorting a Dog and Its Stage");
plt.xticks(rotation=45);
```

The most favorite dog stage is the **doggo,puppo** as on average this stage is favorite of more than 43k persons

```python
stage_retweet = df.groupby('stage')['retweet_count'].mean()
stage_retweet.plot(kind = 'bar', figsize=(10,5), color=(0.2, 0.4, 0.6, 1))
plt.xlabel("Dog Stage")
plt.ylabel("Retweet Count")
plt.title("The Relation Between Retweeting a Dog and Its Stage");
plt.xticks(rotation=45);
```

The most retweeted dog stage is the **doggo,puppo** as on average this stage is retweeted by more than 16k persons

```python
plt.scatter(x=df["rating"], y=df["retweet_count"], alpha=0.5)
plt.figsize = (10,5)
plt.xlabel("Rating")
plt.ylabel("Retweet Count")
plt.title("The Relation Between Retweeting a Dog and Its Rating");
plt.xticks(rotation=45);
```

The chart above shows that the relation between the `rating` and the `retweet_count` is positive, when the dog gets high rating it is retweeted more than low rating dogs

```python
plt.scatter(x=df["rating"], y=df["favorite_count"], alpha=0.5)
plt.figsize = (10,5)
plt.xlabel("Rating")
plt.ylabel("Favorite Count")
plt.title("The Relation Between Favoriting a Dog and Its Rating");
plt.xticks(rotation=45);
```

The chart above shows that the relation between the `rating` and the `favorite_count` is positive, when the dog gets high rating it is retweeted more than low rating dogs

```python
plt.scatter(x=df["favorite_count"], y=df["retweet_count"], alpha=0.5)
plt.figsize = (10,5)
plt.xlabel("Favorite Count")
plt.ylabel("Retweet  Count")
plt.title("The Relation Between Favoriting and Retweeting");
plt.xticks(rotation=45);
```

The chart shows the strong positive relation. The count of people retweets a tweet approximalty equals the half people who likes (or favorite) a tweet.
<br> The following chart zooms to show the relation for the small retweet and favorite counts

```python
ax1 = plt.subplot(212)
plt.scatter(x=df["favorite_count"], y=df["retweet_count"], alpha=0.5)
plt.figsize = (10,10)
plt.xlabel("Favorite Count")
plt.ylabel("Retweet  Count")
plt.title("Zoom in: The Relation Between Favoriting and Retweeting");
ax1.set_xlim(0, 2000)
ax1.set_ylim(0, 1000)
plt.xticks(rotation=45);
```
